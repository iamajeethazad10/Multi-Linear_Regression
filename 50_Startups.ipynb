```python
startups_new = startups.drop(startups.index[[49]],axis=0).reset_index(drop=True)
startups_new
```

```python
model_2 = smf.ols('Profit~RDS+ADMN+MKS',data = startups_new).fit()
```

```python
while model_2.rsquared < 0.99:
    for c in [np.max(c)>1]:
        model_2=smf.ols("Profit~RDS+ADMN+MKS",data=startups_new).fit()
        (c,_)=model_2.get_influence().cooks_distance
        c
        np.argmax(c) , np.max(c)
        startups_new=startups_new.drop(startups_new.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
        startups_new
    else:
        final_model=smf.ols("Profit~RDS+ADMN+MKS",data=startups_new).fit()
        final_model.rsquared , final_model.aic
        print("Thus model accuracy is improved to",final_model.rsquared)
```
